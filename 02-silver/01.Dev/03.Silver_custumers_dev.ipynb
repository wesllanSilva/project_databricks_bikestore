```sql
%python
#criar dicionario path
bronze_map={
    #'tmp_bronze_brands' : f'{bronze_path}/brands/',
    #'tmp_bronze_categories' : f'{bronze_path}/categories/',
    'tmp_bronze_customers' : f'{bronze_path}/customers/',
    #'tmp_bronze_order_items' : f'{bronze_path}/order_items/',
    #'tmp_bronze_orders' : f'{bronze_path}/orders/',
    #'tmp_bronze_products' : f'{bronze_path}/products/',
    #'tmp_bronze_staffs' : f'{bronze_path}/staffs/',
    #'tmp_bronze_stocks' : f'{bronze_path}/stocks/',
    #'tmp_bronze_stores' : f'{bronze_path}/stores/'
}
for view_name, path in bronze_map.items():
    (
        spark.read.format('delta')
        .load(path)
        .createOrReplaceTempView(view_name)
    )
```

```sql
SELECT 
  ct.customer_id
  ,ct.first_name
  ,ct.last_name
  ,ct.phone
  ,ct.email
  ,ct.street
  ,ct.city
FROM tmp_bronze_customers ct
WHERE ct.phone IS NOT NULL AND ct.phone NOT IN('NULL', 'NULL ')
AND ct.email IS NOT NULL AND ct.email NOT IN('NULL', 'NULL ')
```

```sql
%python
df_custumers_silver = spark.sql("""

SELECT 
  ct.customer_id
  ,ct.first_name
  ,ct.last_name
  ,ct.phone
  ,ct.email
  ,ct.street
  ,ct.city
FROM tmp_bronze_customers ct
WHERE ct.phone IS NOT NULL AND ct.phone NOT IN('NULL', 'NULL ')
AND ct.email IS NOT NULL AND ct.email NOT IN('NULL', 'NULL ')

""")
#salvar em delta na silver
df_custumers_silver.write\
    .mode('overwrite')\
    .format('delta')\
    .option('mergeSchema', 'true')\
    .save(f'{silver_path}/customers')
```

```sql
CREATE TABLE IF NOT EXISTS bikestore.logistics.silver_custumers AS(
SELECT 
  ct.customer_id
  ,ct.first_name
  ,ct.last_name
  ,ct.phone
  ,ct.email
  ,ct.street
  ,ct.city
FROM tmp_bronze_customers ct
WHERE ct.phone IS NOT NULL AND ct.phone NOT IN('NULL', 'NULL ')
AND ct.email IS NOT NULL AND ct.email NOT IN('NULL', 'NULL ')
)
```
