```sql
%python
bronze_path   = '/Volumes/bikestore/resource/bronze/'
silver_path   = '/Volumes/bikestore/resource/silver/'
gold_path     = '/Volumes/bikestore/resource/gold/'
resource_path = '/Workspace/Users/wesllan2000@yahoo.com.br/project_databricks_bikestore/resource/origem'
```

```sql
%python
df_gold_sales_ny = spark.sql("""
SELECT 
    shipped_date
    ,round(sum(total_sale),2) as total_sale
 FROM bikestore.logistics.silver_orders
 WHERE state = 'NY'
 AND order_status = 'Delivered'
 GROUP BY shipped_date

 
""")

#salvar em Delta 
df_gold_sales_ny.write\
    .mode('overwrite')\
    .format('delta')\
    .option('mergeSchema', 'true')\
    .save(f'{gold_path}/sales_ny_gold')
```

```sql
CREATE OR REPLACE TABLE bikestore.logistics.sales_ny_gold AS(
  SELECT 
    shipped_date
    ,round(sum(total_sale),2) as total_sale
 FROM bikestore.logistics.silver_orders
 WHERE state = 'NY'
 AND order_status = 'Delivered'
 GROUP BY shipped_date
)
```
